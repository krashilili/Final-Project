```python
#Import Dependencies
import requests
import random
import json
import os
import numpy as np
import pandas as pd
from config import gkey
```

```python
# Import cities file as DataFrame
cities_pd = pd.read_csv("Resources/Dice_US_jobs.csv", encoding = "ISO-8859-1")
cities_pd["Lat"] = ""
cities_pd["Lng"] = ""
cities_pd = cities_pd[cities_pd.City != 'Unspecified']
cities_pd
```

```python
cities_df2 = cities_pd.groupby(['Location']).count()
cities_df3 = pd.DataFrame(cities_df2['City'])
cities_df3.reset_index(inplace=True, level=0)
cities_df3['city_name'] = cities_df3['Location'].apply(lambda city_loc: city_loc.split(',')[0])
cities_df3['state_name'] = cities_df3['Location'].apply(lambda city_loc: city_loc.split(',')[1])
```

```python
# 1- Rename City column to Location Count

cities_df3["Lat"] = ""
cities_df3["Lng"] = ""
cities_df3
```

```python
# params dictionary that will be updated with new city each iteration

params = {"key": gkey}

# Loop through the cities_pd and search for lat/long points for each city

for index, row in cities_df3.iterrows():
        try:
            base_url = "https://maps.googleapis.com/maps/api/geocode/json"

            city = row['city_name']
            state = row['state_name']

            # update address key value
            params['address'] = f"{city},{state}"

            # make request
            cities_lat_lng = requests.get(base_url, params=params)
            print(cities_lat_lng.url)

            # convert to json
            cities_lat_lng = cities_lat_lng.json()

            cities_df3.loc[index, "Lat"] = cities_lat_lng["results"][0]["geometry"]["location"]["lat"]
            cities_df3.loc[index, "Lng"] = cities_lat_lng["results"][0]["geometry"]["location"]["lng"]

        except(IndexError):
            print("Error, skipping to next row")
            
#Visualize to confirm lat lng appear
cities_df3.head()
```

```python
Final_cities_df = cities_df3.sort_values(by='City', ascending=False)
```

```python
Final_cities_df.to_csv('Final_cities_df.csv', encoding='utf-8', index=False)
```
